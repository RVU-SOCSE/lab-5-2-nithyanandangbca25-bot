# shutil module helps you perform common file-management tasks easily.

Common operations:

        Copy files
        
        Move files
        
        Copy folders
        
        Delete folders

        | Function                    | What it Does                                                   |
| --------------------------- | -------------------------------------------------------------- |
| `shutil.copy(src, dst)`     | Copy a **single file** from `src` to `dst`                     |
| `shutil.copytree(src, dst)` | Copy an **entire folder** (including all subfolders and files) |
| `shutil.move(src, dst)`     | Move a **file or folder** to a new location (can also rename)  |
| `shutil.rmtree(path)`       | Delete a **folder and all its contents**                       |
| `shutil.copy2(src, dst)`    | Copy a file **with metadata** (timestamps, permissions)        |
